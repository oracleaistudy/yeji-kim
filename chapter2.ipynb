## 02-1. 훈련 세트와 테스트 세트

#### 지도 학습과 비지도 학습
- **지도 학습** : 훈련하기 위한 데이터와 정답이 필요함
  훈련 데이터 : **입력**(데이터)과 **타깃**(정답)
  각 입력값에 사용된 값 : **특성**
- 머신러닝의 정확한 평가를 위해서는 **테스트 세트**와 **훈련세트**가 따로 준비되어야 함
- 훈련 데이터 : 규칙 학습용
- 테스트 데이터 : 일반화 성능 검증용

-> 테스트 데이터가 아니라 훈련 데이터에만 성능이 좋을 경우 
   과적합(Overfitting)의 문제 발생
